## Creating Tensors

### Reduction operations
These operations reduce the dimensions of a tensor by applying a function (like sum, mean, min, max, etc.) along a specified dimension or over all elements.

Creating a random integer tensor for reduction operations.

```python
e = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
e
```

Calculating the sum of all elements, sum along columns (dim=0), and sum along rows (dim=1).

```python
# sum
print(torch.sum(e))

# sum of columns
print(torch.sum(e, dim=0))

# sum of rows
print(torch.sum(e, dim=1))
```

Calculating the mean of all elements, mean along columns, and mean along rows.

```python
# mean
print(torch.mean(e))

# mean of columns
print(torch.mean(e, dim=0))

# mean of rows
print(torch.mean(e, dim=1))
```

Calculating the median of all elements, median along columns, and median along rows. Note that `torch.median` returns both the values and their indices.

```python
# median
print(torch.median(e))

# median of columns
print(torch.median(e, dim=0))

# median of rows
print(torch.median(e, dim=1))
```

Finding the minimum and maximum values in the tensor.

```python
# min and max
print(torch.min(e))
print(torch.max(e))
```

Calculating the product of all elements, product along columns, and product along rows.

```python
# product
print(torch.prod(e))

# product of columns
print(torch.prod(e, dim=0))

# product of rows
print(torch.prod(e, dim=1))
```

Calculating the standard deviation of all elements, standard deviation along columns, and standard deviation along rows.

```python
# standard deviation
print(torch.std(e))

# standard deviation of columns
print(torch.std(e, dim=0))

# standard deviation of rows
print(torch.std(e, dim=1))
```

Calculating the variance of all elements, variance along columns, and variance along rows.

```python
# variance
print(torch.var(e))

# variance of columns
print(torch.var(e, dim=0))

# variance of rows
print(torch.var(e, dim=1))
```

Finding the index of the maximum value (`argmax`) along different dimensions.

```python
# argmax
print(torch.argmax(e))
print(torch.argmax(e, dim=0))
print(torch.argmax(e, dim=1))
```

Finding the index of the minimum value (`argmin`) along different dimensions.

```python
# argmin
print(torch.argmin(e))
print(torch.argmin(e, dim=0))
print(torch.argmin(e, dim=1))
```

## Inplace Operations
Inplace operations modify the tensor directly without creating a new one, which can save memory. They are often denoted by a trailing underscore (e.g., `add_`).

Creating two random tensors to demonstrate inplace operations.

```python
x = torch.rand(2, 4)
y = torch.rand(2, 4)
print(x)
print(y)
```

Normal addition creates a new tensor, consuming extra memory.

```python
x + y # here it is creating an extra space to save this in memory
```

Using `x.add_(y)` performs addition inplace, modifying `x` directly.

```python
# use inplace op
x.add_(y)
```

```python
x
```

```python
y
```

Applying an inplace sigmoid activation function to `y`.

```python
y.sigmoid_()
```

```python
y
```

## Copy a tensor
This section explains the difference between assigning a tensor to a new variable (which creates a reference) and cloning a tensor (which creates a deep copy).

```python
r = torch.rand(2, 3)
r
```

Creating a sample tensor `r`.

```python
s = r
```

Assigning `r` to `s` creates a reference, so `s` points to the same memory as `r`.

```python
r[0][0] = 399
```

Modifying `r` will also change `s`.

```python
r
```

```python
s
```

Confirming that `r` and `s` have the same memory ID.

```python
id(r) == id(s)
```

Using `clone()` creates a new tensor `s` with a separate copy of the data from `r`.

```python
s = r.clone()
```

Modifying `r` after cloning will not affect `s`.

```python
r[0][0] = 10
```

```python
r
```

```python
s
```

Confirming that `r` and `s` now have different memory IDs.

```python
id(r) == id(s)
```

## Tensor operations on GPU
This section demonstrates how to move tensors to a GPU and perform operations on them, highlighting the potential speedup.

Checking if a CUDA-enabled GPU is available.

```python
torch.cuda.is_available()
```

Setting the device to `cuda` if a GPU is available.

```python
device = torch.device("cuda")
```

Creating a new tensor directly on the GPU.

```python
# creating new tensor on GPU
torch.rand((2,3), device=device)
```

Moving an existing tensor (`s`) from CPU to GPU using the `.to()` method.

```python
# lets move s tensor to GPU
n_t = s.to(device)
```

Benchmarking matrix multiplication performance on CPU versus GPU.

```python
n_t
```

```python
# lets compare CPU vs GPU speed
import time

size = 10000

x = torch.rand(size, size)
y = torch.rand(size, size)

start = time.time()
mat_mul_cpu = torch.matmul(x, y)
end = time.time()

cpu_time = end - start

print(f"CPU time: {cpu_time}")

x_gpu = x.to(device)
y_gpu = y.to(device)

start = time.time()
mat_mul_gpu = torch.matmul(x_gpu, y_gpu)
torch.cuda.synchronize()
end = time.time()

gpu_time = end - start

print(f"GPU time: {gpu_time}")

print(f"Speedup: {cpu_time/gpu_time}")
```

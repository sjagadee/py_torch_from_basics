```python
import torch, torch.nn as nn
```

```python
# create class for model
class Model(nn.Module):

    def __init__(self, no_of_features):
        super().__init__()
        self.linear = nn.Linear(no_of_features, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, feature):
        out = self.linear(feature) # wx + b
        out = self.sigmoid(out)
        return out
```

```python
# create a data set
features = torch.rand(10, 5)

# create the model
model = Model(features.shape[1])

# call model's forward pass
# model.forward(features) # can be used using magic functions in pytorch
model(features)
```

```python
# can check weights
print(model.linear.weight)
print(model.linear.bias)
```

```python
import torch, torch.nn as nn
from torchinfo import summary
```

```python
# Define a simple linear model using nn.Module
class Model(nn.Module):

    # Initialize the model with the number of input features
    def __init__(self, no_of_features):
        super().__init__()
        # Define a linear layer that maps 'no_of_features' to 1 output
        self.linear = nn.Linear(no_of_features, 1)
        # Define a sigmoid activation function
        self.sigmoid = nn.Sigmoid()

    # Define the forward pass of the model
    def forward(self, feature):
        out = self.linear(feature) # Apply linear transformation (wx + b)
        out = self.sigmoid(out)    # Apply sigmoid activation to the output
        return out
```

```python
# Create a dummy dataset with 10 samples and 5 features
features = torch.rand(10, 5)

# Create an instance of the Model
model = Model(features.shape[1]) # Pass the number of features (5) to the model

# Call the model's forward pass to get an output
# The __call__ method of nn.Module automatically invokes the forward method
# model.forward(features) # This is equivalent but less common
output = model(features)
print("Output from Model:\n", output)
```

```python
# Access and print the weights and biases of the linear layers in Model
print("Linear weights:\n", model.linear.weight)
print("Linear bias:\n",model.linear.bias)
```

```python
# Display a summary of the Model architecture
# input_size specifies the shape of a single input batch (batch_size, num_features)
summary(model, input_size=(10, 5))
```

```python
# Define a multi-layer perceptron (MLP) model with one hidden layer
class Model2(nn.Module):

    # Initialize the model with the number of input features
    def __init__(self, no_of_features):
        super().__init__()
        # First linear layer: input features to 3 hidden units
        self.linear1 = nn.Linear(no_of_features, 3)
        # ReLU activation for the hidden layer
        self.relu = nn.ReLU()
        # Second linear layer: 3 hidden units to 1 output
        self.linear2 = nn.Linear(3, 1)
        # Sigmoid activation for the output layer (for binary classification)
        self.sigmoid = nn.Sigmoid()

    # Define the forward pass of the model
    def forward(self, feature):
        out = self.linear1(feature) # Apply first linear transformation
        out = self.relu(out)        # Apply ReLU activation
        out = self.linear2(out)     # Apply second linear transformation
        out = self.sigmoid(out)     # Apply sigmoid activation
        return out
```

```python
# Create a dummy dataset with 10 samples and 5 features
features = torch.rand(10, 5)

# Create an instance of Model2
model = Model2(features.shape[1]) # Pass the number of features (5)

# Call the model's forward pass
output = model(features)
print("Output from Model2:\n", output)
```

```python
# Access and print the weights and biases of the linear layers in Model2
print("Linear 1 weights:\n", model.linear1.weight)
print("Linear 1 bias:\n", model.linear1.bias)
print("Linear 2 weights:\n", model.linear2.weight)
print("Linear 2 bias:\n", model.linear2.bias)
```

```python
# Display a summary of the Model2 architecture
summary(model, input_size=(10, 5))
```

```python
# Define another multi-layer perceptron (MLP) using nn.Sequential
class Model3(nn.Module):

    # Initialize the model with the number of input features
    def __init__(self, no_of_features):
        super().__init__()
        # Define the network as a sequential container of layers
        self.network = nn.Sequential(
            nn.Linear(no_of_features, 3), # First linear layer
            nn.ReLU(),                    # ReLU activation
            nn.Linear(3, 1),              # Second linear layer
            nn.Sigmoid()                  # Sigmoid activation
        )

    # Define the forward pass
    def forward(self, feature):
        out = self.network(feature) # Pass the input through the sequential network
        return out
```

```python
# Create a dummy dataset with 10 samples and 5 features
features = torch.rand(10, 5)

# Create an instance of Model3
model = Model3(features.shape[1]) # Pass the number of features (5)

# Call the model's forward pass
output = model(features)
print("Output from Model3:\n", output)
```

```python
# Access and print the weights and biases of the linear layers in Model3
# When using nn.Sequential, layers are accessed by index from the 'network' attribute.
print("Linear 1 weights:\n", model.network[0].weight)
print("Linear 1 bias:\n", model.network[0].bias)
print("Linear 2 weights:\n", model.network[2].weight)
print("Linear 2 bias:\n", model.network[2].bias)
```

```python
# Display a summary of the Model3 architecture
summary(model, input_size=(10, 5))
```
